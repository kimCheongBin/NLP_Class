# download 함수 만들기

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36"}
```

```python
def download(method="get", url="", headers={}, params={}):
    resp = requests.request(method, 
                            url, 
                            headers=headers,
                            params=params if method=="get" else {},
                            data = params if method=="post" else {}
                           )
    
    if resp.status_code == 200:
        return resp
    
    return None
```

```python
resp = download(url="https://www.google.com/", headers=headers,
               params={'q':'성소'})
```

# 에러 처리 연습
- 지정한 error code를 반환하는 사이트
- https://httpstat.us
- 뒤에 에러코드를 붙이는대로 error를 내준다. ex) /404 => 404 에러 발생
#### 에러코드 설명
- 200 : ok
- 300 : ?
- 400 : 사용자 에러
- 500 : 서버 에러

```python
# error가 났기 떄문에 resp가 None이다. 

resp = download(url="https://httpstat.us/404", headers=headers,
               params={'q':'성소'})
type(resp)
```

```python
# error를 고려해서 함수를 수정하자.
# if문으로 예외를 처리해보자
# 500번대 에러는 잠시 기다린 뒤에 다시 시도하면 되는 경우가 많으므로, 이를 코드에서 추가하자.

import time
import random

def download(method="get", url="", headers={}, params={}, retries=3):
    resp = requests.request(method, 
                            url, 
                            headers=headers,
                            params=params if method=="get" else {},
                            data = params if method=="post" else {}
                           )
    
    if resp.status_code == 200:
        return resp
    else:
        if 500 <= resp.status_code < 600 and retries > 0:
            print("Retries:",retries)
            time.sleep(random.randint(1,5))
            return download(method, url, headers, params, retries-1)
            
        else:
            print(resp.status_code)
            print(resp.reason)
            print(resp.request.headers)
            print(resp.url)

    return resp
```

```python
# 에러가 발생하면 print한다.
# 잘 동작한다.

resp = download('get', url="https://httpstat.us/404", headers=headers,
               params={'q':'성소'})
```

```python
# 500번대 에러가 발생하면 재시도해본다.
# 잘 동작한다.

resp = download('get', url="https://httpstat.us/505", headers=headers,
               params={'q':'성소'})
```

### raise_for_status와 try문으로 조금 더 우아하게 코딩해보자

```python
# response할때 error가 났더라도, 코드에서는 에러를 발생시키지 않는다. 
# error가 난 상태를 response로 받아오기 때문이다.
# raise_for_status()는 response 결과가 에러일 경우에, 코드상 에러를 강제로 발생시키는 메서드이다.

resp = requests.request('get', url="https://httpstat.us/404", headers=headers,
               params={'q':'성소'})

resp.raise_for_status()
```

```python
# raise_for_status 메서드를 이용해서 함수를 수정해보자.

import time
import random

def download(method="get", url="", headers={}, params={}, retries=3):
    try:
        resp = requests.request(method, 
                                url, 
                                headers=headers,
                                params=params if method=="get" else {},
                                data = params if method=="post" else {}
                               )
        resp.raise_for_status()
        
    except requests.exceptions.HTTPError as e:
        if 500 <= resp.status_code < 600 and retries > 0:
            print("Retries:",retries)
            time.sleep(random.randint(1,5))
            return download(method, url, headers, params, retries-1)
            
        else:
            print(resp.status_code)
            print(resp.reason)
            print(resp.request.headers)
            print(resp.url)

    return resp
```

```python
# 잘 동작한다.

resp = download('get', url="https://httpstat.us/404", headers=headers,
               params={'q':'성소'})
```

```python
# 잘 동작한다.

resp = download('get', url="https://httpstat.us/505", headers=headers,
               params={'q':'성소'})
```
